### Precision@K

```python
# Paper definition:
# Precision@k = fraction of relevant games in the recommended set
#               divided by total number of recommended games considered.
#
# Assumption:
# Relevant items are test items with rating >= 7.0

def precision_at_k(recommended_items, relevant_items, k):
    rec_k = recommended_items[:k]
    if k == 0:
        return 0.0

    hits = len(set(rec_k) & set(relevant_items))
    return hits / k


def mean_precision_at_k(recommendations_by_user, relevant_items_by_user, k):
    scores = []

    for user, recs in recommendations_by_user.items():
        relevant_items = relevant_items_by_user.get(user, [])
        scores.append(precision_at_k(recs, relevant_items, k))

    return float(np.mean(scores)) if scores else 0.0
```

### Recall@K

```python
# Paper definition:
# Recall@k = fraction of relevant games in the recommended set
#            divided by the total count of relevant games.
#
# Assumption:
# Relevant items are test items with rating >= 7.0.

def recall_at_k(recommended_items, relevant_items, k):
    relevant_set = set(relevant_items)
    if len(relevant_set) == 0:
        return 0.0

    rec_k = recommended_items[:k]
    hits = len(set(rec_k) & relevant_set)
    return hits / len(relevant_set)


def mean_recall_at_k(recommendations_by_user, relevant_items_by_user, k):
    scores = []

    for user, recs in recommendations_by_user.items():
        relevant_items = relevant_items_by_user.get(user, [])
        scores.append(recall_at_k(recs, relevant_items, k))

    return float(np.mean(scores)) if scores else 0.0
```

NDCG@K

```python
# Paper definition:
# nDCG uses a graded relevance scale and discounts usefulness by rank.
#
# Assumption:
# The paper does not specify the exact graded relevance transformation.
# We therefore use the raw held-out test rating (0-10) as the graded relevance.
#
# Items not present in the user's test set get graded relevance = 0.

def dcg_at_k(recommended_items, graded_relevance_dict, k):
    rec_k = recommended_items[:k]
    dcg = 0.0

    for rank_idx, item in enumerate(rec_k):
        rel = graded_relevance_dict.get(item, 0.0)
        dcg += rel / math.log2(rank_idx + 2)

    return dcg


def ndcg_at_k(recommended_items, graded_relevance_dict, k):
    if len(graded_relevance_dict) == 0:
        return 0.0

    dcg = dcg_at_k(recommended_items, graded_relevance_dict, k)

    ideal_relevances = sorted(graded_relevance_dict.values(), reverse=True)[:k]
    idcg = sum(rel / math.log2(i + 2) for i, rel in enumerate(ideal_relevances))

    if idcg == 0:
        return 0.0

    return dcg / idcg


def mean_ndcg_at_k(recommendations_by_user, graded_relevance_by_user, k):
    scores = []

    for user, recs in recommendations_by_user.items():
        graded_relevance_dict = graded_relevance_by_user.get(user, {})
        scores.append(ndcg_at_k(recs, graded_relevance_dict, k))

    return float(np.mean(scores)) if scores else 0.0
```

### MAP

```python
# Paper definition:
# AP for a user is the average precision over recall calue 0 to 1.
# MAP is the mean of AP across users.
#
# Standard offline implementation:
# We compute AP as the average of Precision@rank at each rank where
# a relevant item is retrieved.
#
# Assumption:
# Relevant items are test items with rating >= 7.0.

def average_precision(recommended_items, relevant_items):
    relevant_set = set(relevant_items)
    if len(relevant_set) == 0:
        return 0.0

    hit_count = 0
    precision_sum = 0.0

    for rank_idx, item in enumerate(recommended_items, start=1):
        if item in relevant_set:
            hit_count += 1
            precision_sum += hit_count / rank_idx

    return precision_sum / len(relevant_set)


def mean_average_precision(recommendations_by_user, relevant_items_by_user):
    scores = []

    for user, recs in recommendations_by_user.items():
        relevant_items = relevant_items_by_user.get(user, [])
        scores.append(average_precision(recs, relevant_items))

    return float(np.mean(scores)) if scores else 0.0
```

### Distance Calculation for Diversity and Novelty

```python
# Faster version - feature vectors + Cached item distance

import numpy as np

# Precompute item to numpy vector maps
mechanics_vectors = {
    item: row.to_numpy(dtype=np.int8)
    for item, row in mechanics_bool_df.iterrows()
}

categories_vectors = {
    item: row.to_numpy(dtype=np.int8)
    for item, row in categories_bool_df.iterrows()
}

# Cache pairwise distances
distance_cache = {}

def item_distance_fast(item_a, item_b, mechanics_vectors, categories_vectors, distance_cache):
    # Use ordered key so (a,b) and (b,a) share the same cache entry
    key = tuple(sorted((item_a, item_b)))

    if key in distance_cache:
        return distance_cache[key]

    if item_a not in mechanics_vectors or item_b not in mechanics_vectors:
        return None
    if item_a not in categories_vectors or item_b not in categories_vectors:
        return None

    mech_a = mechanics_vectors[item_a]
    mech_b = mechanics_vectors[item_b]
    cat_a = categories_vectors[item_a]
    cat_b = categories_vectors[item_b]

    diff_mechanics = np.mean(mech_a != mech_b)
    diff_categories = np.mean(cat_a != cat_b)

    d = (diff_mechanics + diff_categories) / 2.0
    distance_cache[key] = float(d)
    return float(d)
```

```python
# The paper defines the distance between two games as:
#
#   ((#DiffMechanics / #TotalMechanics) +
#    (#DiffCategories / #TotalCategories)) / 2
#
# Mechanics and Categories must be stored separately as boolean arrays.
#
# Assumption:
# If either item is missing from the feature tables, we return None
# and skip that pair in later calculations.

import numpy as np

def item_distance(item_a, item_b, mechanics_bool_df, categories_bool_df):
    if item_a not in mechanics_bool_df.index or item_b not in mechanics_bool_df.index:
        return None
    if item_a not in categories_bool_df.index or item_b not in categories_bool_df.index:
        return None

    mech_a = mechanics_bool_df.loc[item_a].astype(int).values
    mech_b = mechanics_bool_df.loc[item_b].astype(int).values

    cat_a = categories_bool_df.loc[item_a].astype(int).values
    cat_b = categories_bool_df.loc[item_b].astype(int).values

    total_mechanics = len(mech_a)
    total_categories = len(cat_a)

    if total_mechanics == 0 or total_categories == 0:
        return None

    diff_mechanics = np.sum(mech_a != mech_b)
    diff_categories = np.sum(cat_a != cat_b)

    distance = ((diff_mechanics / total_mechanics) +
                (diff_categories / total_categories)) / 2.0

    return float(distance)
```

### Diversity

```python
# faster version - with cache

def diversity_for_user_fast(recommended_items, mechanics_vectors, categories_vectors, distance_cache):
    items = list(recommended_items)
    k = len(items)

    if k == 0:
        return 0.0

    total_distance = 0.0
    count = 0

    for i in range(k):
        for j in range(k):
            d = item_distance_fast(items[i], items[j], mechanics_vectors, categories_vectors, distance_cache)
            if d is not None:
                total_distance += d
                count += 1

    return total_distance / count if count > 0 else 0.0


def mean_diversity_fast(recommendations_by_user, mechanics_vectors, categories_vectors, distance_cache, k=None):
    scores = []

    for user, recs in recommendations_by_user.items():
        rec_list = recs[:k] if k is not None else recs
        scores.append(
            diversity_for_user_fast(rec_list, mechanics_vectors, categories_vectors, distance_cache)
        )

    return float(np.mean(scores)) if scores else 0.0
```

```python
# The paper defines Diversity as the pairwise differences between
# items of the same set of recommendations:
#
#   Diff_Diversity = sum_i ( sum_j |X_i - X_j| )

def diversity_for_user(recommended_items, mechanics_bool_df, categories_bool_df):
    items = list(recommended_items)
    k = len(items)

    if k == 0:
        return 0.0

    total_distance = 0.0
    count = 0

    for i in range(k):
        for j in range(k):
            d = item_distance(items[i], items[j], mechanics_bool_df, categories_bool_df)
            if d is not None:
                total_distance += d
                count += 1

    return total_distance / count if count > 0 else 0.0


def mean_diversity(recommendations_by_user, mechanics_bool_df, categories_bool_df, k=None):
    scores = []

    for user, recs in recommendations_by_user.items():
        rec_list = recs[:k] if k is not None else recs
        scores.append(diversity_for_user(rec_list, mechanics_bool_df, categories_bool_df))

    return float(np.mean(scores)) if scores else 0.0
```

### Novelty

```python
# faaster version

def novelty_for_user_fast(recommended_items, liked_items, mechanics_vectors, categories_vectors, distance_cache):
    recs = list(recommended_items)
    liked = list(liked_items)

    k = len(recs)
    l = len(liked)

    if k == 0 or l == 0:
        return 0.0

    total_distance = 0.0
    count = 0

    for i in range(k):
        for j in range(l):
            d = item_distance_fast(recs[i], liked[j], mechanics_vectors, categories_vectors, distance_cache)
            if d is not None:
                total_distance += d
                count += 1

    return total_distance / count if count > 0 else 0.0


def mean_novelty_fast(recommendations_by_user, liked_items_by_user, mechanics_vectors, categories_vectors, distance_cache, k=None):
    scores = []

    for user, recs in recommendations_by_user.items():
        rec_list = recs[:k] if k is not None else recs
        liked_items = liked_items_by_user.get(user, [])
        scores.append(
            novelty_for_user_fast(rec_list, liked_items, mechanics_vectors, categories_vectors, distance_cache)
        )

    return float(np.mean(scores)) if scores else 0.0
```

```python
# The paper defines Novelty as the pairwise differences between:
# - X = the recommendation list
# - Y = the user's historically liked items
#
#   Diff_Novelty = sum_i sum_j |X_i - Y_j|
#
# This means:
# - compare every recommended item against every liked item
# - use the same paper distance as in Eq. (1)
#
# Assumption:
# "Liked items" -> rating >= 7.0

def novelty_for_user(recommended_items, liked_items, mechanics_bool_df, categories_bool_df):
    recs = list(recommended_items)
    liked = list(liked_items)

    k = len(recs)
    l = len(liked)

    if k == 0 or l == 0:
        return 0.0

    total_distance = 0.0
    count = 0

    for i in range(k):
        for j in range(l):
            d = item_distance(recs[i], liked[j], mechanics_bool_df, categories_bool_df)
            if d is not None:
                total_distance += d
                count += 1

    return total_distance / count if count > 0 else 0.0


def mean_novelty(recommendations_by_user, liked_items_by_user, mechanics_bool_df, categories_bool_df, k=None):
    scores = []

    for user, recs in recommendations_by_user.items():
        rec_list = recs[:k] if k is not None else recs
        liked_items = liked_items_by_user.get(user, [])
        scores.append(novelty_for_user(rec_list, liked_items, mechanics_bool_df, categories_bool_df))

    return float(np.mean(scores)) if scores else 0.0
```

```python
# Make a recommendation list for one user. Any model can be used.
def recommend_for_user(model, fallback, trainset, user_id, user_seen_items, all_candidate_items, top_n=100, transform_estimate=None):
    """
    uses the model to estimate "predictions" on every item for the user.
    Predictions are sorted in descending order and the top top_n items are returned.

    Arguments
    model: an object that implements the estimate(inner_uid, inner_iid) function.
        estimate() should return a float, but other types are ok as long as the transform_estimate argument is also specified.
        If estimate cannot gnerate a prediction, it should raise PredictionImpossible.
    fallback: The value to use as the prediction if estimate raises PredictionImpossible.
    trainset: the surprise trainset.
    user_id: The raw user id for whom recommendations are to be made.
    user_seen_items: a list of items the user has already seen.
    all_candidate_items: all possible items to be recommended to any user.
    top_n: (optional) how many items to recommend. Default 100.
    transform_estimate: (optional) if it is not None, this is called on the output of estimate to get a single float value score.
    """
    # convert inputed user to a string if not in str format already
    user_id = str(user_id)

    # if the user is not known to the trainset, then no prediction can be made
    try:
        inner_uid = trainset.to_inner_uid(user_id)
    except ValueError:
        return []

    # grab the items that user already knows for exclusion.
    seen_items = user_seen_items.get(user_id, set())

    # score all candidate items not already rated in train set
    scored_items = []
    for raw_iid in all_candidate_items:
        if raw_iid in seen_items:
            continue

        # if an item is unknown don't even try to rate it
        try:
            inner_iid = trainset.to_inner_iid(str(raw_iid))
        except ValueError:
            continue

        # if an item cannot be predicted for some other reason, return the fallback
        try:
            score = model.estimate(inner_uid, inner_iid)
            if not (transform_estimate is None):
                score = transform_estimate(score)
        except PredictionImpossible:
            score = fallback

        scored_items.append((raw_iid, score))

    scored_items.sort(key=lambda x: x[1], reverse=True) # descending order

    return [item for item, score in scored_items[:top_n]]
```

```python

import os
import pickle
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# Build user - item matrix

# Assuming The paper's autoencoder is user-based:
# input = one user's full rating vector over items
# output = reconstructed rating vector
#
# Missing ratings are filled with 0 in the input matrix.
# Loss is computed only on observed entries using a mask.

ae_users = sorted(train_surprise_norm['Username'].astype(str).unique())
ae_items = sorted(train_surprise_norm['BGGId'].astype(str).unique())

user_to_idx = {u: i for i, u in enumerate(ae_users)}
item_to_idx = {it: j for j, it in enumerate(ae_items)}

num_users = len(ae_users)
num_items = len(ae_items)

R_train = np.zeros((num_users, num_items), dtype=np.float32)
M_train = np.zeros((num_users, num_items), dtype=np.float32)

for row in train_surprise_norm.itertuples(index=False):
    u = user_to_idx[str(row.Username)]
    i = item_to_idx[str(row.BGGId)]
    R_train[u, i] = float(row.Rating_norm)
    M_train[u, i] = 1.0

print("Train matrix shape:", R_train.shape)
print("Observed train ratings:", int(M_train.sum()))
print("Normalized rating range:", R_train[M_train == 1].min(), R_train[M_train == 1].max())
```

```python
# dataset for pytorch

# - batch size: 128

class UserRatingsDataset(Dataset):
    def __init__(self, ratings_matrix, mask_matrix):
        self.X = torch.tensor(ratings_matrix, dtype=torch.float32)
        self.M = torch.tensor(mask_matrix, dtype=torch.float32)

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.M[idx]

train_dataset = UserRatingsDataset(R_train, M_train)
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
```

```python
# Classic autoencoder model

# Paper settings:
# - hidden layer: 1024
# - activation: ReLU

class ClassicAutoencoder(nn.Module):
    def __init__(self, num_items, hidden_dim=1024):
        super().__init__()
        self.encoder = nn.Linear(num_items, hidden_dim)
        self.decoder = nn.Linear(hidden_dim, num_items)
        self.relu = nn.ReLU()

    def forward(self, x):
        h = self.relu(self.encoder(x))
        out = self.decoder(h)
        return out
```

```python
# Training

# Paper settings:
# - epochs: 50
# - regularization: 0.01
# - optimizer: Adadelta
#
# Assumption:
# Loss = masked MSE over observed entries only.

def masked_mse_loss(pred, target, mask):
    diff = (pred - target) * mask
    denom = mask.sum()
    if denom.item() == 0:
        return torch.tensor(0.0, device=pred.device, requires_grad=True)
    return diff.pow(2).sum() / denom

def train_classic_autoencoder(model, train_loader, epochs=50, weight_decay=0.01):
    model = model.to(device)
    optimizer = optim.Adadelta(model.parameters(), weight_decay=weight_decay)
    loss_history = []

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0.0

        for x, m in train_loader:
            x = x.to(device)
            m = m.to(device)

            optimizer.zero_grad()
            pred = model(x)
            loss = masked_mse_loss(pred, x, m)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_epoch_loss = epoch_loss / len(train_loader)
        loss_history.append(avg_epoch_loss)
        print(f"Epoch {epoch+1}/{epochs} - loss: {avg_epoch_loss:.6f}")

    return model, loss_history
```

```python
classic_ae_model = ClassicAutoencoder(num_items=num_items, hidden_dim=1024)
classic_ae_model, classic_ae_loss_history = train_classic_autoencoder(
    classic_ae_model,
    train_loader,
    epochs=50,
    weight_decay=0.01
)

model_name = "autoencoder"
model_dir = os.path.join(save_dir, model_name)
os.makedirs(model_dir, exist_ok=True)

# save the model
with open(os.path.join(model_dir, 'model.pkl'), 'wb') as file:
    pickle.dump(classic_ae_model, file)

# save the model artifacts needed for reuse
classic_ae_artifacts = {
    'ae_users': ae_users,
    'ae_items': ae_items,
    'user_to_idx': user_to_idx,
    'item_to_idx': item_to_idx,
    'loss_history': classic_ae_loss_history
}

with open(os.path.join(model_dir, 'artifacts.pkl'), 'wb') as file:
    pickle.dump(classic_ae_artifacts, file)

print("Classic autoencoder model and artifacts saved.")
```

```python
#
# Generate Recommendations
#

# For each evaluation user:
# - predict full reconstructed normalized rating vector
# - de-normalize by adding back the user's train mean
# - remove seen items
# - rank remaining items
# - keep top 100

classic_ae_model.eval()
classic_ae_recommendations_by_user = {}

with torch.no_grad():
    for user in tqdm(evaluation_users):
        if user not in user_to_idx:
            classic_ae_recommendations_by_user[user] = []
            continue

        uidx = user_to_idx[user]

        x = torch.tensor(R_train[uidx], dtype=torch.float32, device=device).unsqueeze(0)
        pred_norm = classic_ae_model(x).squeeze(0).cpu().numpy()

        # add back user mean to get predicted raw ratings
        user_mean = float(user_means.loc[user])
        pred = pred_norm + user_mean

        seen = user_seen_items.get(user, set())

        scored_items = []
        for j, item in enumerate(ae_items):
            if item in seen:
                continue
            scored_items.append((item, float(pred[j])))

        scored_items.sort(key=lambda x: x[1], reverse=True)
        classic_ae_recommendations_by_user[user] = [item for item, score in scored_items[:100]]

# save recommendations
with open(os.path.join(model_dir, 'recommendations.pkl'), 'wb') as file:
    pickle.dump(classic_ae_recommendations_by_user, file)

print("Classic autoencoder recommendations saved.")
```

```python
import os
import pickle
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model_name = "user_cdae_denoising_autoencoder"
model_dir = os.path.join(save_dir, model_name)
os.makedirs(model_dir, exist_ok=True)
```

```python
# Build matrix

# Input = one user's full normalized rating vector over items
# Output = reconstructed normalized rating vector over items

ae_users = sorted(train_surprise_norm['Username'].astype(str).unique())
ae_items = sorted(train_surprise_norm['BGGId'].astype(str).unique())

user_to_idx = {u: i for i, u in enumerate(ae_users)}
item_to_idx = {it: j for j, it in enumerate(ae_items)}

num_users = len(ae_users)
num_items = len(ae_items)

R_train = np.zeros((num_users, num_items), dtype=np.float32)
M_train = np.zeros((num_users, num_items), dtype=np.float32)

for row in train_surprise_norm.itertuples(index=False):
    u = user_to_idx[str(row.Username)]
    i = item_to_idx[str(row.BGGId)]
    R_train[u, i] = float(row.Rating_norm)
    M_train[u, i] = 1.0

print("Train matrix shape:", R_train.shape)
print("Observed train ratings:", int(M_train.sum()))
print("Normalized rating range:", R_train[M_train == 1].min(), R_train[M_train == 1].max())
```

```python
# dataset

class UserRatingsWithIdDataset(Dataset):
    def __init__(self, ratings_matrix, mask_matrix):
        self.X = torch.tensor(ratings_matrix, dtype=torch.float32)
        self.M = torch.tensor(mask_matrix, dtype=torch.float32)
        self.user_ids = torch.arange(self.X.shape[0], dtype=torch.long)

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.user_ids[idx], self.X[idx], self.M[idx]

train_dataset = UserRatingsWithIdDataset(R_train, M_train)
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
```

```python
# Input corrpution func

def corrupt_input(x, mask, corruption_rate=0.1):
    # only corrupt observed entries
    noise = torch.bernoulli(torch.full_like(x, 1 - corruption_rate))
    effective_noise = torch.where(mask > 0, noise, torch.ones_like(noise))

    x_corrupted = x * effective_noise

    # unbiased scaling for retained observed entries
    scale = 1.0 / (1.0 - corruption_rate)
    x_corrupted = torch.where((mask > 0) & (effective_noise > 0), x_corrupted * scale, x_corrupted)

    return x_corrupted
```

```python
# Model

class UserCDAE(nn.Module):
    def __init__(self, num_users, num_items, hidden_dim=1024):
        super().__init__()
        self.encoder = nn.Linear(num_items, hidden_dim)
        self.user_embedding = nn.Embedding(num_users, hidden_dim)
        self.decoder = nn.Linear(hidden_dim, num_items)
        self.relu = nn.ReLU()

    def forward(self, user_ids, x):
        h = self.encoder(x) + self.user_embedding(user_ids)
        h = self.relu(h)
        out = self.decoder(h)
        return out
```

```python
# Train

# Paper settings:
# - epochs = 100
# - regularization = 0.01
# - optimizer = Adadelta
#
# Assumption:
# Loss = masked MSE over observed entries only.

def masked_mse_loss(pred, target, mask):
    diff = (pred - target) * mask
    denom = mask.sum()
    if denom.item() == 0:
        return torch.tensor(0.0, device=pred.device, requires_grad=True)
    return diff.pow(2).sum() / denom

def train_user_cdae(model, train_loader, epochs=100, weight_decay=0.01, corruption_rate=0.1):
    model = model.to(device)
    optimizer = optim.Adadelta(model.parameters(), weight_decay=weight_decay)
    loss_history = []

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0.0

        for user_ids, x, m in train_loader:
            user_ids = user_ids.to(device)
            x = x.to(device)
            m = m.to(device)

            x_corrupted = corrupt_input(x, m, corruption_rate=corruption_rate)

            optimizer.zero_grad()
            pred = model(user_ids, x_corrupted)
            loss = masked_mse_loss(pred, x, m)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_epoch_loss = epoch_loss / len(train_loader)
        loss_history.append(avg_epoch_loss)
        print(f"Epoch {epoch+1}/{epochs} - loss: {avg_epoch_loss:.6f}")

    return model, loss_history
```

```python
# Fit

user_cdae_model = UserCDAE(num_users=num_users, num_items=num_items, hidden_dim=1024)

user_cdae_model, user_cdae_loss_history = train_user_cdae(
    user_cdae_model,
    train_loader,
    epochs=100,
    weight_decay=0.01,
    corruption_rate=0.1
)

with open(os.path.join(model_dir, 'model.pkl'), 'wb') as file:
    pickle.dump(user_cdae_model, file)

user_cdae_artifacts = {
    'ae_users': ae_users,
    'ae_items': ae_items,
    'user_to_idx': user_to_idx,
    'item_to_idx': item_to_idx,
    'loss_history': user_cdae_loss_history
}

with open(os.path.join(model_dir, 'artifacts.pkl'), 'wb') as file:
    pickle.dump(user_cdae_artifacts, file)

print("User-based CDAE denoising autoencoder model and artifacts saved.")
print("Saved to:", model_dir)
```

```python
# Generating recommendations

user_cdae_model.eval()
user_cdae_recommendations_by_user = {}

with torch.no_grad():
    for user in tqdm(evaluation_users):
        if user not in user_to_idx:
            user_cdae_recommendations_by_user[user] = []
            continue

        uidx = user_to_idx[user]

        x = torch.tensor(R_train[uidx], dtype=torch.float32, device=device).unsqueeze(0)
        user_ids = torch.tensor([uidx], dtype=torch.long, device=device)

        pred_norm = user_cdae_model(user_ids, x).squeeze(0).cpu().numpy()

        # de-normalize
        user_mean = float(user_means.loc[user])
        pred = pred_norm + user_mean

        seen = user_seen_items.get(user, set())

        scored_items = []
        for j, item in enumerate(ae_items):
            if item in seen:
                continue
            scored_items.append((item, float(pred[j])))

        scored_items.sort(key=lambda x: x[1], reverse=True)
        user_cdae_recommendations_by_user[user] = [item for item, score in scored_items[:100]]

with open(os.path.join(model_dir, 'recommendations.pkl'), 'wb') as file:
    pickle.dump(user_cdae_recommendations_by_user, file)

print("User-based CDAE recommendations saved.")
```

```python


import os
import pickle
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model_name = "denoising_autoencoder_item_gaussian"
model_dir = os.path.join(save_dir, model_name)
os.makedirs(model_dir, exist_ok=True)
```

```python
# Train
# Settings:
# - epochs = 100
# - weight decay = 0.01
# - optimizer = Adadelta


def masked_mse_loss(pred, target, mask):
    diff = (pred - target) * mask
    denom = mask.sum()
    if denom.item() == 0:
        return torch.tensor(0.0, device=pred.device, requires_grad=True)
    return diff.pow(2).sum() / denom

def train_item_denoising_autoencoder(model, train_loader, epochs=100, weight_decay=0.01):
    model = model.to(device)
    optimizer = optim.Adadelta(model.parameters(), weight_decay=weight_decay)
    loss_history = []

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0.0

        for x, m in train_loader:
            x = x.to(device)
            m = m.to(device)

            optimizer.zero_grad()
            pred = model(x, mask=m)
            loss = masked_mse_loss(pred, x, m)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_epoch_loss = epoch_loss / len(train_loader)
        loss_history.append(avg_epoch_loss)
        print(f"Epoch {epoch+1}/100 - loss: {avg_epoch_loss:.6f}")

    return model, loss_history
```

## Variational AutoEncoders

```python


import os
import pickle
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model_name = "variational_autoencoder_user"
model_dir = os.path.join(save_dir, model_name)
os.makedirs(model_dir, exist_ok=True)
```

```python
# build matrix

vae_users = sorted(train_surprise_norm['Username'].astype(str).unique())
vae_items = sorted(train_surprise_norm['BGGId'].astype(str).unique())

user_to_idx = {u: i for i, u in enumerate(vae_users)}
item_to_idx = {it: j for j, it in enumerate(vae_items)}

num_users = len(vae_users)
num_items = len(vae_items)

R_train_norm = np.zeros((num_users, num_items), dtype=np.float32)
M_train = np.zeros((num_users, num_items), dtype=np.float32)

for row in train_surprise_norm.itertuples(index=False):
    u = user_to_idx[str(row.Username)]
    i = item_to_idx[str(row.BGGId)]
    R_train_norm[u, i] = float(row.Rating_norm)
    M_train[u, i] = 1.0

print("Train matrix shape:", R_train_norm.shape)
print("Observed train ratings:", int(M_train.sum()))
print("Observed normalized range:", R_train_norm[M_train == 1].min(), R_train_norm[M_train == 1].max())
```

```python
# Scale ratings to [0,1] for BCE
# Assumption:
# The board game paper says BCE is used, but does not explain how
# normalized explicit ratings are transformed into BCE-compatible targets.
# We use min-max scaling computed from observed training normalized ratings only.

observed_vals = R_train_norm[M_train == 1]
norm_min = float(observed_vals.min())
norm_max = float(observed_vals.max())

def scale_to_unit_interval(x, x_min, x_max):
    return (x - x_min) / (x_max - x_min)

def unscale_from_unit_interval(x_scaled, x_min, x_max):
    return x_scaled * (x_max - x_min) + x_min

R_train_bce = np.zeros_like(R_train_norm, dtype=np.float32)
R_train_bce[M_train == 1] = scale_to_unit_interval(observed_vals, norm_min, norm_max)

print("Scaled observed range:", R_train_bce[M_train == 1].min(), R_train_bce[M_train == 1].max())
print("norm_min:", norm_min)
print("norm_max:", norm_max)
```

```python
# dataset

# Paper settings:
# batch size = 64

class UserRatingsDataset(Dataset):
    def __init__(self, ratings_matrix, mask_matrix):
        self.X = torch.tensor(ratings_matrix, dtype=torch.float32)
        self.M = torch.tensor(mask_matrix, dtype=torch.float32)

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.M[idx]

train_dataset = UserRatingsDataset(R_train_bce, M_train)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

print("Number of users in train dataset:", len(train_dataset))
```

```python
# model

# Paper settings:
# - hidden size = 256
#
# Assumption:
# The board game paper says "256 hidden nodes" but does not separately
# specify latent dimension, so we use latent_dim = 256 as well.

class UserVAE(nn.Module):
    def __init__(self, num_items, hidden_dim=256, latent_dim=256):
        super().__init__()

        self.fc1 = nn.Linear(num_items, hidden_dim)
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)

        self.fc2 = nn.Linear(latent_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, num_items)

        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def encode(self, x):
        h1 = self.relu(self.fc1(x))
        mu = self.fc_mu(h1)
        logvar = self.fc_logvar(h1)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h2 = self.relu(self.fc2(z))
        return self.sigmoid(self.fc3(h2))

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        recon = self.decode(z)
        return recon, mu, logvar
```

```python
# Paper settings:
# - BCE loss rather than MSE
# - epochs = 50
# - batch size = 64
# - regularization = 1e-05
#
# VAE loss:
# BCE reconstruction term + KL divergence.
# Weight decay is set to 1e-05

def vae_loss_function(recon_x, x, mu, logvar, mask):
    # BCE only on observed entries
    bce = nn.functional.binary_cross_entropy(
        recon_x * mask,
        x * mask,
        reduction='sum'
    )

    denom = mask.sum()
    if denom.item() == 0:
        bce = torch.tensor(0.0, device=x.device)
    else:
        bce = bce / denom

    # KL divergence
    kld = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())

    return bce + kld, bce, kld

def train_user_vae(model, train_loader, epochs=50, weight_decay=1e-05):
    model = model.to(device)
    optimizer = optim.Adadelta(model.parameters(), weight_decay=weight_decay)

    loss_history = []

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0.0

        for x, m in train_loader:
            x = x.to(device)
            m = m.to(device)

            optimizer.zero_grad()

            recon, mu, logvar = model(x)
            loss, bce, kld = vae_loss_function(recon, x, mu, logvar, m)

            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_epoch_loss = epoch_loss / len(train_loader)
        loss_history.append(avg_epoch_loss)
        print(f"Epoch {epoch+1}/50 - loss: {avg_epoch_loss:.6f}")

    return model, loss_history
```

```python

user_vae_model = UserVAE(
    num_items=num_items,
    hidden_dim=256,
    latent_dim=256
)

user_vae_model, user_vae_loss_history = train_user_vae(
    user_vae_model,
    train_loader,
    epochs=50,
    weight_decay=1e-05
)

with open(os.path.join(model_dir, 'model.pkl'), 'wb') as file:
    pickle.dump(user_vae_model, file)

artifacts = {
    'vae_users': vae_users,
    'vae_items': vae_items,
    'user_to_idx': user_to_idx,
    'item_to_idx': item_to_idx,
    'loss_history': user_vae_loss_history,
    'norm_min': norm_min,
    'norm_max': norm_max
}

with open(os.path.join(model_dir, 'artifacts.pkl'), 'wb') as file:
    pickle.dump(artifacts, file)

print("User VAE model and artifacts saved.")
```

```python
# Recommendation
# - model input is scaled normalized ratings in [0,1]
# - model output is predicted scaled normalized ratings in [0,1]
# - convert back to normalized scale
# - then add back user_mean to get raw predicted ratings
# - then rank unseen items

user_vae_model.eval()
user_vae_recommendations_by_user = {}

with torch.no_grad():
    for user in tqdm(evaluation_users):
        if user not in user_to_idx:
            user_vae_recommendations_by_user[user] = []
            continue

        uidx = user_to_idx[user]

        x = torch.tensor(
            R_train_bce[uidx],
            dtype=torch.float32,
            device=device
        ).unsqueeze(0)

        # use posterior mean for prediction
        mu, logvar = user_vae_model.encode(x)
        pred_scaled = user_vae_model.decode(mu).squeeze(0).cpu().numpy()

        # convert back to normalized scale
        pred_norm = unscale_from_unit_interval(pred_scaled, norm_min, norm_max)

        # add back user mean to return to raw rating scale
        user_mean = float(user_means.loc[user])
        pred = pred_norm + user_mean

        seen = user_seen_items.get(user, set())

        scored_items = []
        for j, item in enumerate(vae_items):
            if item in seen:
                continue
            scored_items.append((item, float(pred[j])))

        scored_items.sort(key=lambda x: x[1], reverse=True)

        user_vae_recommendations_by_user[user] = [
            item for item, score in scored_items[:100]
        ]

with open(os.path.join(model_dir, 'recommendations.pkl'), 'wb') as file:
    pickle.dump(user_vae_recommendations_by_user, file)

print("User VAE recommendations saved.")
```

```python

user_vae_metrics_df = evaluate_all_metrics_fast(
    recommendations_by_user=user_vae_recommendations_by_user,
    relevant_items_by_user=relevant_items_by_user,
    graded_relevance_by_user=graded_relevance_by_user,
    liked_items_by_user=liked_items_by_user,
    mechanics_vectors=mechanics_vectors,
    categories_vectors=categories_vectors,
    distance_cache=distance_cache
)

display(user_vae_metrics_df)

with open(os.path.join(model_dir, 'metrics.pkl'), 'wb') as file:
    pickle.dump(user_vae_metrics_df, file)

print("User VAE metrics saved.")
```

```python

import os
import pickle
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm.notebook import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model_name = "variational_autoencoder_item"
model_dir = os.path.join(save_dir, model_name)
os.makedirs(model_dir, exist_ok=True)
```

```python
# Build matrix

item_vae_users = sorted(train_surprise_norm['Username'].astype(str).unique())
item_vae_items = sorted(train_surprise_norm['BGGId'].astype(str).unique())

item_user_to_idx = {u: i for i, u in enumerate(item_vae_users)}
item_item_to_idx = {it: j for j, it in enumerate(item_vae_items)}

num_users_item_vae = len(item_vae_users)
num_items_item_vae = len(item_vae_items)

R_item_train_norm = np.zeros((num_items_item_vae, num_users_item_vae), dtype=np.float32)
M_item_train = np.zeros((num_items_item_vae, num_users_item_vae), dtype=np.float32)

for row in train_surprise_norm.itertuples(index=False):
    u = item_user_to_idx[str(row.Username)]
    i = item_item_to_idx[str(row.BGGId)]
    R_item_train_norm[i, u] = float(row.Rating_norm)
    M_item_train[i, u] = 1.0

print("Item matrix shape:", R_item_train_norm.shape)
print("Observed ratings:", int(M_item_train.sum()))
print("Observed normalized range:", R_item_train_norm[M_item_train == 1].min(), R_item_train_norm[M_item_train == 1].max())
```

```python
# Scale ratings to [0,1] for BCE

observed_vals = R_item_train_norm[M_item_train == 1]
norm_min = float(observed_vals.min())
norm_max = float(observed_vals.max())

def scale_to_unit_interval(x, x_min, x_max):
    return (x - x_min) / (x_max - x_min)

def unscale_from_unit_interval(x_scaled, x_min, x_max):
    return x_scaled * (x_max - x_min) + x_min

R_item_train_bce = np.zeros_like(R_item_train_norm, dtype=np.float32)
R_item_train_bce[M_item_train == 1] = scale_to_unit_interval(
    observed_vals, norm_min, norm_max
)

print("Scaled observed range:", R_item_train_bce[M_item_train == 1].min(), R_item_train_bce[M_item_train == 1].max())
print("norm_min:", norm_min)
print("norm_max:", norm_max)
```

```python
# dataset

# Paper settings:
# batch size = 64

class ItemRatingsDataset(Dataset):
    def __init__(self, ratings_matrix, mask_matrix):
        self.X = torch.tensor(ratings_matrix, dtype=torch.float32)
        self.M = torch.tensor(mask_matrix, dtype=torch.float32)

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.M[idx]

item_train_dataset = ItemRatingsDataset(R_item_train_bce, M_item_train)
item_train_loader = DataLoader(item_train_dataset, batch_size=64, shuffle=True)

print("Number of items in train dataset:", len(item_train_dataset))
```

```python
# model

# Paper settings:
# - hidden size = 256
# Assumption : latent_dim = 256


class ItemVAE(nn.Module):
    def __init__(self, num_users, hidden_dim=256, latent_dim=256):
        super().__init__()

        self.fc1 = nn.Linear(num_users, hidden_dim)
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)

        self.fc2 = nn.Linear(latent_dim, hidden_dim)
        self.fc3 = nn.Linear(hidden_dim, num_users)

        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def encode(self, x):
        h1 = self.relu(self.fc1(x))
        mu = self.fc_mu(h1)
        logvar = self.fc_logvar(h1)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z):
        h2 = self.relu(self.fc2(z))
        return self.sigmoid(self.fc3(h2))

    def forward(self, x):
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        recon = self.decode(z)
        return recon, mu, logvar
```

```python
# Paper settings:
# - BCE loss rather than MSE
# - epochs = 50
# - batch size = 64
# - regularization = 1e-05
#
# VAE loss:
# BCE reconstruction term + KL divergence

def vae_loss_function(recon_x, x, mu, logvar, mask):
    bce = nn.functional.binary_cross_entropy(
        recon_x * mask,
        x * mask,
        reduction='sum'
    )

    denom = mask.sum()
    if denom.item() == 0:
        bce = torch.tensor(0.0, device=x.device)
    else:
        bce = bce / denom

    kld = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())

    return bce + kld, bce, kld

def train_item_vae(model, item_train_loader, epochs=50, weight_decay=1e-05):
    model = model.to(device)
    optimizer = optim.Adadelta(model.parameters(), weight_decay=weight_decay)

    loss_history = []

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0.0

        for x, m in item_train_loader:
            x = x.to(device)
            m = m.to(device)

            optimizer.zero_grad()

            recon, mu, logvar = model(x)
            loss, bce, kld = vae_loss_function(recon, x, mu, logvar, m)

            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_epoch_loss = epoch_loss / len(item_train_loader)
        loss_history.append(avg_epoch_loss)
        print(f"Epoch {epoch+1}/50 - loss: {avg_epoch_loss:.6f}")

    return model, loss_history
```

```python


item_vae_model = ItemVAE(
    num_users=num_users_item_vae,
    hidden_dim=256,
    latent_dim=200
)

item_vae_model, item_vae_loss_history = train_item_vae(
    item_vae_model,
    item_train_loader,
    epochs=50,
    weight_decay=1e-05
)

with open(os.path.join(model_dir, 'model.pkl'), 'wb') as file:
    pickle.dump(item_vae_model, file)

artifacts = {
    'item_vae_users': item_vae_users,
    'item_vae_items': item_vae_items,
    'item_user_to_idx': item_user_to_idx,
    'item_item_to_idx': item_item_to_idx,
    'loss_history': item_vae_loss_history,
    'norm_min': norm_min,
    'norm_max': norm_max
}

with open(os.path.join(model_dir, 'artifacts.pkl'), 'wb') as file:
    pickle.dump(artifacts, file)

print("Item VAE model and artifacts saved.")
```

## Matrix Factorization

```python
import os
import pickle
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm.auto import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
mf_users = sorted(train_surprise_norm['Username'].astype(str).unique())
mf_items = sorted(train_surprise_norm['BGGId'].astype(str).unique())

user_to_idx_mf = {u: i for i, u in enumerate(mf_users)}
item_to_idx_mf = {it: j for j, it in enumerate(mf_items)}

num_users_mf = len(mf_users)
num_items_mf = len(mf_items)

train_df_mf = train_surprise_norm.copy()
train_df_mf['user_idx'] = train_df_mf['Username'].astype(str).map(user_to_idx_mf)
train_df_mf['item_idx'] = train_df_mf['BGGId'].astype(str).map(item_to_idx_mf)
train_df_mf['rating'] = train_df_mf['Rating_norm'].astype(np.float32)

print(train_df_mf)
```

```python
class MFDataset(Dataset):
    def __init__(self, df):
        self.user_idx = torch.tensor(df['user_idx'].values, dtype=torch.long)
        self.item_idx = torch.tensor(df['item_idx'].values, dtype=torch.long)
        self.ratings  = torch.tensor(df['rating'].values,   dtype=torch.float32)

    def __len__(self):
        return len(self.ratings)

    def __getitem__(self, idx):
        return self.user_idx[idx], self.item_idx[idx], self.ratings[idx]

mf_train_dataset = MFDataset(train_df_mf)
mf_train_loader  = DataLoader(mf_train_dataset, batch_size=1024, shuffle=True)
```

```python
class MatrixFactorization(nn.Module):
    def __init__(self, num_users, num_items, emb_dim=5):
        super().__init__()
        self.user_emb = nn.Embedding(num_users, emb_dim)
        self.item_emb = nn.Embedding(num_items, emb_dim)

        nn.init.normal_(self.user_emb.weight, mean=0, std=0.1)
        nn.init.normal_(self.item_emb.weight, mean=0, std=0.1)

    def forward(self, user_idx, item_idx):
        u = self.user_emb(user_idx)
        i = self.item_emb(item_idx)
        return (u * i).sum(dim=1)
```

```python
def train_mf(model, train_loader, epochs=20):
    model = model.to(device)

    optimizer = torch.optim.Adam(
        model.parameters(),
        lr=1e-3,
        weight_decay=1e-5
    )

    loss_fn = nn.MSELoss()
    loss_history = []

    for epoch in tqdm(range(epochs), desc="Epochs"):
        model.train()
        total_loss = 0.0

        for u, i, r in train_loader:
            u = u.to(device)
            i = i.to(device)
            r = r.to(device)

            optimizer.zero_grad()
            pred = model(u, i)
            loss = loss_fn(pred, r)
            loss.backward()
            optimizer.step()

            total_loss += loss.item() * len(r)

        avg_loss = total_loss / len(train_loader.dataset)
        loss_history.append(avg_loss)

        print(f"Epoch {epoch+1}/{epochs} - Loss: {avg_loss:.6f}")

    return model, loss_history
```

```python
mf_model = MatrixFactorization(num_users=num_users_mf, num_items=num_items_mf, emb_dim=64)
mf_model, mf_loss_history = train_mf(mf_model, mf_train_loader, epochs=50)

model_name = "matrix_factorization"
model_dir_mf = os.path.join(save_dir, model_name)
os.makedirs(model_dir_mf, exist_ok=True)

with open(os.path.join(model_dir_mf, 'model.pkl'), 'wb') as f:
  pickle.dump(mf_model, f)

mf_artifacts = {
    'mf_users':       mf_users,
    'mf_items':       mf_items,
    'user_to_idx_mf': user_to_idx_mf,
    'item_to_idx_mf': item_to_idx_mf,
    'loss_history':   mf_loss_history
}

with open(os.path.join(model_dir_mf, 'artifacts.pkl'), 'wb') as f:
    pickle.dump(mf_artifacts, f)


print("Matrix Factorization model and artifacts saved.")
```

```python
mf_recommendations_by_user = {}
mf_model.eval()

with torch.no_grad():
    item_emb = mf_model.item_emb(
        torch.arange(num_items_mf, device=device)
    )

    for user in tqdm(evaluation_users):
        if user not in user_to_idx_mf:
            mf_recommendations_by_user[user] = []
            continue

        uidx = user_to_idx_mf[user]
        u_tensor = torch.tensor([uidx], device=device)
        user_emb = mf_model.user_emb(u_tensor)

        scores = (item_emb @ user_emb.T).squeeze(1)
        scores = scores.cpu().numpy()

        if user in user_means.index:
            user_mean = float(user_means.loc[user])
        else:
            user_mean = 0.0

        scores = scores + user_mean

        seen = user_seen_items.get(user, set())

        scored_items = []
        for j, item in enumerate(mf_items):
            if item not in seen:
                scored_items.append((item, float(scores[j])))

        scored_items.sort(key=lambda x: x[1], reverse=True)

        mf_recommendations_by_user[user] = [
            item for item, _ in scored_items[:100]
        ]

with open(os.path.join(model_dir_mf, 'recommendations.pkl'), 'wb') as file:
    pickle.dump(mf_recommendations_by_user, file)

print("Matrix Factorization recommendations saved.")
```

## KNN Content filtering

**Weighted KNN**

Ref: https://www.geeksforgeeks.org/machine-learning/weighted-k-nn/

We apply a **weighted KNN approach**, where each neighbor is scaled using a smoothed inverse distance function.

**Inverse Distance Weighting**

w = 1 / (1+d)

where:
- (d): distance between items  
- (w): weight assigned to the neighbor  

Closer neighbors receive higher weights, while farther neighbors contribute less to the final score.

## IDF-Based


Ref: https://www.geeksforgeeks.org/machine-learning/understanding-tf-idf-term-frequency-inverse-document-frequency/

For inverse document frequency (IDF), we use a **smoothed version** to avoid division by zero.

**IDF Formula**

idf = log((N + 1) / (df + 1)) + 1

where:
- \(N\): total number of games  
- \(df\): number of games containing the feature  

```python
mechanics_features_df = mechanics.copy()
categories_features_df = subcategories.copy()

mechanics_features_df['BGGId'] = mechanics_features_df['BGGId'].astype(str)
categories_features_df['BGGId'] = categories_features_df['BGGId'].astype(str)

content_features_df = mechanics_features_df.merge(
    categories_features_df,
    on='BGGId',
    how='left'
)

content_features_df = content_features_df.fillna(0)
content_features_df = content_features_df.drop_duplicates(subset='BGGId')

print(f"Content feature table shape: {content_features_df.shape}")
```

```python
item_ids = content_features_df['BGGId'].tolist()
item_to_idx = {item: i for i, item in enumerate(item_ids)}

feature_cols = [col for col in content_features_df.columns if col != "BGGId"]
X = content_features_df[feature_cols].values.astype(np.float32)

num_items = X.shape[0]
num_features = X.shape[1]

print(f"Number of items: {num_items}")
print(f"Number of content features: {num_features}")

items_in_feature = X.sum(axis=0)

idf = np.log((num_items + 1) / (items_in_feature + 1)) + 1

X_idf = X * idf
```

```python
item_sim = X @ X_idf.T

np.fill_diagonal(item_sim, 0.0)
```

```python
TOP_N = 100
idf_recommendations_by_user = {}

for user in tqdm(evaluation_users):
    seen = user_seen_items.get(user, set())
    liked_items = liked_items_by_user.get(user, [])
    liked_items = [item for item in liked_items if item in item_to_idx]

    if not liked_items:
        idf_recommendations_by_user[user] = []
        continue

    liked_idx = [item_to_idx[item] for item in liked_items]

    score_accum = item_sim[liked_idx].sum(axis=0)

    seen_idx = [item_to_idx[item] for item in seen if item in item_to_idx]
    score_accum[seen_idx] = 0.0

    top_idx = np.argsort(-score_accum)

    recommendation = []

    for idx in top_idx:
        if score_accum[idx] <= 0:
            continue
        recommendation.append(item_ids[idx])

        if len(recommendation) >= TOP_N:
            break

    idf_recommendations_by_user[user] = recommendation

print("IDF-based recommendations created.")

```

```python
def create_edge_index(df, num_users):
  #convert users and items into one index space
  #items index are shift by num_users to avoid collision with user index
  users = torch.tensor(df["userId"].values, dtype=torch.long)
  items = torch.tensor(df["itemId"].values, dtype=torch.long) + num_users

  edge_index = torch.stack([users, items], dim=0)

  edge_index = torch.cat([edge_index, edge_index.flip(0)], dim=1)

  return edge_index
```

```python
class LightGCN(torch.nn.Module):
  def __init__(self, num_users, num_items, embedding_dim=64, num_layers=3):
    super().__init__()

    self.num_users = num_users
    self.num_items = num_items
    self.num_nodes = num_users + num_items
    self.num_layers = num_layers

    self.embedding = torch.nn.Embedding(self.num_nodes, embedding_dim)
    torch.nn.init.normal_(self.embedding.weight, std=0.1)

    self.convs = torch.nn.ModuleList()

    # create k layers of LGC propagation
    for layer in range(num_layers):
      conv_layer = LGConv()
      self.convs.append(conv_layer)

  def forward(self, edge_index):
    emb_0 = self.embedding.weight

    all_embeddings = [emb_0]
    emb_k = emb_0

    for conv in self.convs:
      emb_k = conv(emb_k, edge_index)
      all_embeddings.append(emb_k)

    emb_final = torch.stack(all_embeddings, dim=0).mean(dim=0)

    user_embedding = emb_final[:self.num_users]
    item_embedding = emb_final[self.num_users:]

    users_emb_0 = emb_0[:self.num_users]
    items_emb_0 = emb_0[self.num_users:]

    return user_embedding, users_emb_0, item_embedding, items_emb_0
```

```python
def bpr_loss(
    users_emb_final,
    users_emb_0,
    pos_items_emb_final,
    pos_items_emb_0,
    neg_items_emb_final,
    neg_items_emb_0,
    lambda_val,
):
    pos_scores = torch.sum(users_emb_final * pos_items_emb_final, dim=-1)

    neg_scores = torch.sum(users_emb_final * neg_items_emb_final, dim=-1)

    ranking_loss = -torch.mean(torch.nn.functional.logsigmoid(pos_scores - neg_scores))


    reg_loss = (
        users_emb_0.norm(2).pow(2)
        + pos_items_emb_0.norm(2).pow(2)
        + neg_items_emb_0.norm(2).pow(2)
    ) / users_emb_0.size(0)

    loss = ranking_loss + lambda_val * reg_loss

    return loss
```

```python
def get_user_positive_items(edge_index, num_users):
  user_pos_items = {}

  mask = edge_index[0] < num_users
  users = edge_index[0][mask]
  items = edge_index[1][mask] - num_users
  for user, item in zip(users.tolist(), items.tolist()):
    if user not in user_pos_items:
      user_pos_items[user] = []

    user_pos_items[user].append(item)

  return user_pos_items
```

```python
def sample_batch(edge_index, batch_size, num_users, num_items, user_pos_items):
  mask = edge_index[0] < num_users
  user_nodes = edge_index[0][mask]
  item_nodes = edge_index[1][mask] - num_users

  idx = torch.randint(0, user_nodes.size(0), (batch_size,))

  users = user_nodes[idx]
  pos_items = item_nodes[idx]

  neg_items = []
  for u in users.tolist():
    neg = torch.randint(0, num_items, (1,)).item()

    while neg in user_pos_items[u]:
      neg = torch.randint(0, num_items, (1,)).item()

    neg_items.append(neg)

  neg_items = torch.tensor(neg_items, dtype=torch.long)

  return users, pos_items, neg_items
```

```python
def get_embs_for_bpr(model, edge_index, batch_size, num_users, num_items, user_pos_items, device):
  users_emb_final, users_emb_0, items_emb_final, items_emb_0 = model(edge_index)

  users, pos_items, neg_items = sample_batch(
    edge_index.cpu(),
    batch_size,
    num_users,
    num_items,
    user_pos_items
  )

  users = users.to(device)
  pos_items = pos_items.to(device)
  neg_items = neg_items.to(device)

  return (
    users_emb_final[users],
    users_emb_0[users],
    items_emb_final[pos_items],
    items_emb_0[pos_items],
    items_emb_final[neg_items],
    items_emb_0[neg_items],
  )
```

```python
def train_lightgcn(train_df, num_users, num_items,
                   learning_rate=0.001, iterations=1000,
                   batch_size=1024, device=device, LAMBDA=1e-4):
    train_edge_index = create_edge_index(train_df, num_users).to(device)
    user_pos_items = get_user_positive_items(train_edge_index.cpu(), num_users)

    model = LightGCN(num_users, num_items).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

    train_losses = []

    for step in range(iterations):
        model.train()

        (
            users_emb_final,
            users_emb_0,
            pos_items_emb_final,
            pos_items_emb_0,
            neg_items_emb_final,
            neg_items_emb_0,
        ) = get_embs_for_bpr(
            model,
            train_edge_index,
            batch_size,
            num_users,
            num_items,
            user_pos_items,
            device,
        )

        loss = bpr_loss(
            users_emb_final,
            users_emb_0,
            pos_items_emb_final,
            pos_items_emb_0,
            neg_items_emb_final,
            neg_items_emb_0,
            LAMBDA,
        )

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())

        if step % 200 == 0:
            print(f'[Iteration  {step}/{iterations}] loss: {loss.item():.5f}')

    return model, train_edge_index, user_pos_items, train_losses
```

```python
class HybridAutoEncoder(nn.Module):
    def __init__(self, num_items, num_cb_features, hidden_dim=1024):
        super().__init__()
        self.encoder = nn.Linear(num_items + num_cb_features, hidden_dim)
        self.decoder = nn.Linear(hidden_dim, num_items)
        self.relu = nn.ReLU()

    def forward(self, x):
        h = self.relu(self.encoder(x))
        out = self.decoder(h)
        return out
```

```python
class HybridDataset(Dataset):
    def __init__(self, user_profiles, num_cb_features, mask_matrix):
        # make the autoencoder input
        self.X = torch.tensor(user_profiles, dtype=torch.float32)

        # grab just the ratings part
        self.R = torch.tensor(user_profiles[:,:-num_cb_features], dtype=torch.float32)

        # the mask
        self.M = torch.tensor(mask_matrix, dtype=torch.float32)

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx], self.R[idx], self.M[idx]
```

```python
def masked_mse_loss(pred, target, mask):
    diff = (pred - target) * mask
    denom = mask.sum()
    if denom.item() == 0:
        return torch.tensor(0.0, device=pred.device, requires_grad=True)
    return diff.pow(2).sum() / denom
```

```python
class HybridRecommender:
    def __init__(self, hidden_nodes, epochs, batch_size, regularization, learning_rate, normalized_ratings):
        self.hidden_nodes = hidden_nodes
        self.epochs = epochs
        self.batch_size = batch_size
        self.regularization = regularization
        self.learning_rate = learning_rate
        self.normalized_ratings = normalized_ratings

        # initialize important values to None
        self.hybrid_ae = None
        self.user_profiles = None
        self.user_to_row = None
        self.item_to_col = None

    def compute_user_profiles(self):
        print("computing user profiles")
        # compute the ratings matrix and a mask for evaluation
        rating_profiles_ret = self.compute_rating_profiles_and_mask()

        # compute a content based profile for each user
        content_profiles = self.compute_content_profiles()

        # combine the profiles into a single matrix that can be indexed by a Dataset
        rating_profiles = rating_profiles_ret['rating_profiles']
        user_profiles = self.combine_profiles(rating_profiles, content_profiles)

        # return user_profiles
        print("done computing user profiles")
        return {
            "user_profiles": user_profiles,
            "num_cb_features": content_profiles.shape[1],
            "mask": rating_profiles_ret['mask'],
            "user_to_idx": rating_profiles_ret['user_to_idx'],
            "item_to_idx": rating_profiles_ret['item_to_idx'],
            "idx_to_user": rating_profiles_ret['idx_to_user'],
            "idx_to_item": rating_profiles_ret['idx_to_item'],
        }

    def compute_rating_profiles_and_mask(self):
        print("  computing rating profiles and mask")
        # generate index translation data structures
        user_to_idx = {u: i for i, u in enumerate(self.ae_users)}
        item_to_idx = {it: j for j, it in enumerate(self.ae_items)}

        idx_to_user = {i: u for i, u in enumerate(self.ae_users)}
        idx_to_item = {j: it for j, it in enumerate(self.ae_items)}

        # save index translation data structures
        self.user_to_idx = user_to_idx
        self.item_to_idx = item_to_idx
        self.idx_to_user = idx_to_user
        self.idx_to_item = idx_to_item

        # initialize ratings matrix and the evaluation mask
        rating_profiles = np.zeros((self.num_users, self.num_items), dtype=np.float32)
        mask = np.zeros((self.num_users, self.num_items), dtype=np.float32)

        # populate ratings matrix with user ratings and the evaluation mask with rated items
        for row in self.train_data.itertuples(index=False):
            u = user_to_idx[str(row.Username)]
            i = item_to_idx[str(row.BGGId)]
            if self.normalized_ratings:
                rating_profiles[u, i] = float(row.Rating_norm)
            else:
                rating_profiles[u, i] = float(row.Rating)
            mask[u, i] = 1.0

        # print("Train matrix shape:", rating_profiles.shape)
        # print("Observed train ratings:", int(mask.sum()))
        # print("Rating range:", rating_profiles[mask == 1].min(), rating_profiles[mask == 1].max())

        return {
            "rating_profiles": rating_profiles,
            "mask": mask,
            "user_to_idx": user_to_idx,
            "item_to_idx": item_to_idx,
            "idx_to_user": idx_to_user,
            "idx_to_item": idx_to_item,
        }

    def compute_content_profiles(self):
        print("  computing content based user profiles based on liked items")
        # compute three CB matricies where row is user and column is a CB feature
        self.categories_profiles = self.compute_categories_profiles()
        self.mechanics_profiles = self.compute_mechanics_profiles()
        self.other_attribute_profiles = self.compute_other_attribute_profiles()

        # stack horizontally
        # print(self.categories_profiles.shape)
        # print(self.mechanics_profiles.shape)
        # print(self.other_attribute_profiles.shape)

        # print(self.categories_profiles[:5])
        # print(self.mechanics_profiles[:5])
        # print(self.other_attribute_profiles[:5])
        content_profiles = np.hstack((self.categories_profiles, self.mechanics_profiles, self.other_attribute_profiles))

        return content_profiles

    def compute_categories_profiles(self):
        print("    computing categories TF-IDF content-based profiles")
        # store column names
        self.categories_columns = [col for col in categories_bool_df if col != 'BGGId']

        # init categories counts matrix
        category_counts = np.zeros((self.num_users, len(self.categories_columns)), dtype='int32')

        # get categories counts per user
        for user in self.ae_users:
            # get the user's idx
            user_idx = self.user_to_idx[user]

            # grab the user's liked items
            if user in liked_items_by_user:
                liked_items = liked_items_by_user[user]
            else:
                category_counts[user_idx] = np.zeros((len(self.categories_columns)))
                continue

            # grab the rows of the categories df that corresponsds to those items
            user_cats = categories_bool_df[categories_bool_df.index.isin(liked_items)]

            # renove the bggid column and convert to numpy matrix
            user_cats = user_cats[self.categories_columns].to_numpy()

            # get the counts by summing vertically
            user_cats_counts = np.sum(user_cats, axis=0)

            # populate the user's row in the category counts
            category_counts[user_idx] = user_cats_counts

        # compute TFIDF on category counts
        cat_tfidf = TfidfTransformer()
        categories_profiles = cat_tfidf.fit_transform(category_counts).toarray()
        return categories_profiles

    def compute_mechanics_profiles(self):
        print("    computing mechanics TF-IDF content-based profiles")
        # store column names
        self.mechanics_columns = [col for col in mechanics_bool_df if col != 'BGGId']

        # init mechanics counts matrix
        mechanics_counts = np.zeros((self.num_users, len(self.mechanics_columns)), dtype='int32')

        # get mechanics counts per user
        for user in self.ae_users:
            # get the user's idx
            user_idx = self.user_to_idx[user]

            # grab the user's liked items
            if user in liked_items_by_user:
                liked_items = liked_items_by_user[user]
            else:
                mechanics_counts[user_idx] = np.zeros((len(self.mechanics_columns)))
                continue

            # grab the rows of the mechanics df that corresponsds to those items
            user_mechanics = mechanics_bool_df[mechanics_bool_df.index.isin(liked_items)]

            # renove the bggid column and convert to numpy matrix
            user_mechanics = user_mechanics[self.mechanics_columns].to_numpy()

            # get the counts by summing vertically
            user_mechanics_counts = np.sum(user_mechanics, axis=0)

            # populate the user's row in the mechanics counts
            mechanics_counts[user_idx] = user_mechanics_counts

        # compute TFIDF on category counts
        mechanics_tfidf = TfidfTransformer()
        mechanics_profiles = mechanics_tfidf.fit_transform(mechanics_counts).toarray()
        return mechanics_profiles

    def compute_other_attribute_profiles(self):
        print("    computing other attributes content-based profiles")
        # store column names
        self.other_attribute_columns = ['MfgPlaytime', 'GameWeight', 'MinPlayers', 'MfgAgeRec']

        # init other attributes means matrix
        other_attribute_means = np.zeros((self.num_users, len(self.other_attribute_columns)), dtype='float32')

        # get other attribute values per user
        for user in self.ae_users:
            # get the user's idx
            user_idx = self.user_to_idx[user]

            # grab the user's liked items
            if user in liked_items_by_user:
                liked_items = liked_items_by_user[user]
            else:
                other_attribute_means[user_idx] = np.array([0.0,0.0,0.0,0.0])
                continue

            # grab the rows of the games df that corresponsds to those items
            user_other_attributes = games[games['BGGId'].isin(liked_items)]

            # renove the bggid column and irrelavant others and convert to numpy matrix
            user_other_attributes = user_other_attributes[self.other_attribute_columns].to_numpy()

            # get the means by averaging vertically
            user_other_attribute_means = np.mean(user_other_attributes, axis=0)

            # populate the user's row in the mechanics counts
            other_attribute_means[user_idx] = user_other_attribute_means

        return other_attribute_means

    def combine_profiles(self, rating_profiles, content_profiles):
        print("  combining ratings and content-based profiles")
        # ratings profile is a num_users by num_items matrix
        # content_profiles is a num_users by self.num_cb_features
        combined_profiles = np.hstack((rating_profiles, content_profiles))
        return combined_profiles

    def fit(self, train_data):
        # store training data and basic info
        self.train_data = train_data
        self.ae_users = sorted(self.train_data['Username'].astype(str).unique())
        self.ae_items = sorted(self.train_data['BGGId'].astype(str).unique())
        self.num_users = len(self.ae_users)
        self.num_items = len(self.ae_items)

        # calculate and store the user profiles, mask, num cb features, and index translation
        user_profiles_ret = self.compute_user_profiles()
        self.user_profiles = user_profiles_ret['user_profiles']
        self.num_cb_features = user_profiles_ret['num_cb_features']
        self.mask = user_profiles_ret['mask']
        self.user_to_idx = user_profiles_ret['user_to_idx']
        self.item_to_idx = user_profiles_ret['item_to_idx']
        self.idx_to_user = user_profiles_ret['idx_to_user']
        self.idx_to_item = user_profiles_ret['idx_to_item']

        # initialize a hybrid data set and Dataloader from the user profiles
        self.hybrid_dataset = HybridDataset(user_profiles=self.user_profiles, num_cb_features=self.num_cb_features, mask_matrix=self.mask)
        self.hybrid_dataloader =  DataLoader(self.hybrid_dataset, batch_size=self.batch_size, shuffle=True)

        # initialize a hybrid autoencoder and send to device
        self.hybrid_autoencoder = HybridAutoEncoder(num_items=self.num_items, num_cb_features=self.num_cb_features, hidden_dim=self.hidden_nodes)
        self.hybrid_autoencoder.to(device)

        # define the optimizer
        self.optimizer = optim.Adadelta(self.hybrid_autoencoder.parameters(), weight_decay=self.regularization, lr=self.learning_rate)

        # train the model, printing loss at each epoch
        self.loss_history = []
        for epoch in tqdm(range(self.epochs)):
            self.hybrid_autoencoder.train()
            epoch_loss = 0.0

            for x, r, m in self.hybrid_dataloader:
                x = x.to(device)
                r = r.to(device)
                m = m.to(device)

                self.optimizer.zero_grad()
                pred = self.hybrid_autoencoder(x)
                loss = masked_mse_loss(pred, r, m)
                loss.backward()
                self.optimizer.step()

                epoch_loss += loss.item()

            avg_epoch_loss = epoch_loss / len(self.hybrid_dataloader)
            self.loss_history.append(avg_epoch_loss)
            print(f"Epoch {epoch+1}/{self.epochs} - loss: {avg_epoch_loss:.6f}")

    def recommend(self):
        print("generating recommendations for users")
        self.hybrid_autoencoder.eval()

        self.recs_by_user = {}
        with torch.no_grad():
            for user in tqdm(evaluation_users):
                # check for unknown user
                if user not in self.user_to_idx:
                    self.recs_by_user[user] = []
                    continue

                # convert to user index
                uidx = self.user_to_idx[user]

                # get scores for all items for this user
                x = torch.tensor(self.user_profiles[uidx], dtype=torch.float32, device=device).unsqueeze(0)
                pred_norm = self.hybrid_autoencoder(x).squeeze(0).cpu().numpy()

                # add back user mean to get predicted raw ratings
                user_mean = float(user_means.loc[user])
                pred = pred_norm + user_mean

                # get the items this user has seen before to exclude them from recommendation
                seen = user_seen_items.get(user, set())

                # convert from item index to BGGId
                scored_items = []
                for j, item in enumerate(self.ae_items):
                    if item in seen:
                        continue
                    scored_items.append((item, float(pred[j])))

                # sort by predicted rating and store top 100 items
                scored_items.sort(key=lambda x: x[1], reverse=True)
                self.recs_by_user[user] = [item for item, score in scored_items[:100]]

        return self.recs_by_user

    def save(self, save_dir):
        # save the model
        torch.save(self.hybrid_autoencoder, os.path.join(save_dir, "hybrid_autoencoder_model.pth"))

        # save yourself
        with open(os.path.join(save_dir, "hybrid_recommender.pkl"), 'wb') as f:
            pickle.dump(self, f)

    def load_model(self, save_dir):
        # load the model
        self.hybrid_autoencoder = torch.load(os.path.join(save_dir, "hybrid_autoencoder_model.pth"))
        self.hybrid_autoencoder.eval()

    @classmethod
    def load_recommender(cls, save_dir):
        """
        Usage
        # saving
        recommender = HybridRecommender(...)
        recommender.fit(...)
        recommender.save("./models")

        # loading
        loaded_recommender = HybridRecommender.load_recommender("./models")
        loaded_recommender.load_model("./models")
        """
        path = os.path.join(save_dir, "hybrid_recommender.pkl")
        with open(path, 'rb') as f:
            return pickle.load(f)
```

```python
# Batched recommendation for tuning

def generate_user_ae_recommendations_batched(
    model,
    R_train,
    evaluation_users,
    user_to_idx,
    ae_items,
    item_to_idx,
    user_seen_items,
    user_means,
    batch_size=256,
    top_n=100
):
    model.eval()
    model = model.to(device)

    recommendations_by_user = {}
    known_users = [user for user in evaluation_users if user in user_to_idx]

    with torch.no_grad():
        for start in tqdm(range(0, len(known_users), batch_size)):
            batch_users = known_users[start:start + batch_size]
            batch_indices = [user_to_idx[user] for user in batch_users]

            x = torch.tensor(
                R_train[batch_indices],
                dtype=torch.float32,
                device=device
            )

            preds = model(x)

            for row_idx, user in enumerate(batch_users):
                preds[row_idx] += float(user_means.loc[user])

            for row_idx, user in enumerate(batch_users):
                seen_items = user_seen_items.get(user, set())

                for item in seen_items:
                    if item in item_to_idx:
                        preds[row_idx, item_to_idx[item]] = -float("inf")

            _, top_indices = torch.topk(preds, k=top_n, dim=1)
            top_indices = top_indices.cpu().numpy()

            for row_idx, user in enumerate(batch_users):
                recommendations_by_user[user] = [
                    ae_items[item_idx] for item_idx in top_indices[row_idx]
                ]

    for user in evaluation_users:
        if user not in recommendations_by_user:
            recommendations_by_user[user] = []

    return recommendations_by_user
```

```python
# Check performance with Adadelta and Adam Optimizer

def train_autoencoder_with_optimizer(
    model,
    train_loader,
    epochs=50,
    weight_decay=0.1,
    optimizer_name="Adadelta",
    lr=1e-3
):
    model = model.to(device)

    if optimizer_name == "Adadelta":
        optimizer = optim.Adadelta(model.parameters(), weight_decay=weight_decay)
    elif optimizer_name == "Adam":
        optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    else:
        raise ValueError("optimizer_name must be 'Adadelta' or 'Adam'")

    loss_history = []

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0.0

        for x, m in train_loader:
            x = x.to(device)
            m = m.to(device)

            optimizer.zero_grad()
            pred = model(x)
            loss = masked_mse_loss(pred, x, m)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        avg_epoch_loss = epoch_loss / len(train_loader)
        loss_history.append(avg_epoch_loss)

        print(f"Epoch {epoch+1}/{epochs} - loss: {avg_epoch_loss:.6f}")

    return model, loss_history
```
